## Univariate

```python
# remove those with low variance
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
fs = VarianceThreshold(threshold= 1)
```

```python
fs.fit(df)
```

```python
fs.get_support()
```

```python
new_df = df.iloc[:,fs.get_support()]
```
